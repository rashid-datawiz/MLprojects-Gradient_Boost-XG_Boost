```python
#We need to impute Null values in the dataframe we will use simple imputers and 
# we will have two categories here one with discrete null values and one with continuous null values
discrete_null_columns = ['Gender','Married','Dependents','Self_Employed','Loan_Amount_Term','Credit_History']
continous_null_columns = ['LoanAmount']
```

```python
OHE_columns=["Self_Employed",'Property_Area','Gender']
standard_scaler = ["ApplicantIncome","CoapplicantIncome","LoanAmount"]
function_transformer = ["Loan_Amount_Term"] # creating a variable for custom function
label_encoder=["Married"]
pass_through = ['Credit_History'] # should remain as it is
```

```python
#we need to define ordinalencoder seperately for every column as we have to define the order of categories
dependents_enc = OrdinalEncoder(categories=[['0', '1', '2', '3+']])
education_enc = OrdinalEncoder(categories=[['Not Graduate','Graduate']])
```

### Note: Column Transformer is one functionality in which we mention which column will take which transformation
- Before we go to Transformation, those initialization of which column will take which transformation is crucial

### Steps of preprocesing for features:
- LabelEncoder for Married,Credit_histroy
- OrdinalEncoder for Dependents,Education
- One hot encoder for Self Employed,Property_area,Gender
- FucntionTransformer for Loan_Amount_term
- Standard scaler for Applicant Income , CoapplicantIncome and LoanAmount
- Credit histroy will pass through
### Steps of preprocessing for Label
- LabelEncoder for Loan_Status (Approved)

```python
# Function transformer is used whenever we are using custom functions, and to use that custom function we need FuncctionTransformer
custom=FunctionTransformer(divide_by_12)
```

```python
class ModifiedLabelEncoder(LabelEncoder):
    def fit_transform(self, y, *args, **kwargs):
        return super().fit_transform(y).reshape(-1, 1)

    def transform(self, y, *args, **kwargs):
        return super().transform(y).reshape(-1, 1)
```

```python
no_transform=FunctionTransformer(same)
```

```python
preprocessor=ColumnTransformer([("OHE columns",OneHotEncoder(),OHE_columns),
                                ("Label_encoder",ModifiedLabelEncoder(),"Married"),
                                ("standard_scaler",StandardScaler(),standard_scaler),
                                ("custom",custom,function_transformer),
                                ("pass through",no_transform,["Credit_History"]),
                                ("ordinal dependents",dependents_enc,['Dependents']),
                                ("ordinal education",education_enc,['Education']),
                                ("Label Label encoder",ModifiedLabelEncoder(),["Loan_Status (Approved)"])])
```
